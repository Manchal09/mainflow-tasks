```python
# A)Data Loading
import pandas as pd
file_path="/content/student-mat.csv"
student_data=pd.read_csv(file_path, sep=';')
print(student_data.head())
# B) Data Exploration
student_data.isnull().sum()
student_data.dtypes
student_data.shape
# C) Data Cleaning
student_data.drop_duplicates()
# D) Data Analysis Questions
# 1
avg_G3=student_data['G3'].mean()
print(f"Average Final Grade in Maths (G3): {avg_G3}")
#2
score_above_15=student_data[student_data['G3']>15].shape[0]
print(f" No. of students scoring above 15 : {score_above_15}")
#3 Correlation between study time (study time) and the final grade (G3)
import numpy as np
correlation=np.corrcoef(student_data["studytime"] ,student_data["G3"])[0,1]
print(f" Correlation between study time (study time) and the final grade (G3) : {correlation}")
#4 Which gender has a higher average final grade
avg_G3_gender=student_data.groupby("sex")["G3"].mean()
print(avg_G3_gender)
```

```python
# E) Data Visualization
# Histogram of final grades (G3)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.histplot(student_data["G3"],bins=10 ,kde=True)
plt.title("Distribution of Final Grades [G3] ")
plt.xlabel("final grade (G3)")
plt.ylabel("frequency")
plt.show()
```

```python
# Scatter plot between study time (study time) and final grade (G3)
plt.figure(figsize=(8,5))
sns.scatterplot(x=student_data["studytime"],y=student_data["G3"])
plt.title("Study Time VS Final grades")
plt.xlabel("Study time (hours per week)")
plt.ylabel("final grade [G3]")
plt.show()
```

```python
# Bar chart comparing the average scores of male and female students
plt.figure(figsize=(8,5))
sns.barplot(x=avg_G3_gender.index,y=avg_G3_gender.values)
plt.title("Average final grade [G3] by gender")
plt.xlabel("Gender")
plt.ylabel("Average final grade [G3]")
plt.show()
```
