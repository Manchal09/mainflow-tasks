```python
#Import liberaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Loading dataset
file_path= "/content/Global_Superstore(CSV).csv"
sales_data = pd.read_csv(file_path)
sales_data=pd.read_csv(file_path)
print(sales_data.head())   #Display first 5 rows of the dataset

#Basic information about the dataset
print("\n Dataset Information")
sales_data.info()

#Satistical Summary of numerical columns
print("\n Satistical Summary")
display(sales_data.describe())

#Check for Duplicates
duplicates=sales_data.duplicated().sum()
print(f"Number of duplicate Rows: {duplicates}")

#Remove Duplicates
sales_data=sales_data.drop_duplicates()

```

```python
#Handle missing values
print(f"Missing values before cleaning : \n {sales_data.isnull().sum()}")

#Fill missing numerical values with column mean
sales_data.fillna(sales_data.select_dtypes(include=['number']).mean(), inplace=True)  #mean only to numerical columns
print(f"Missing values after cleaning : \n {sales_data.isnull().sum()}")

import pandas as pd
#Convert 'Date' column to datetime format
sales_data['Order Date']=pd.to_datetime(sales_data['Order Date'])
sales_data['Ship Date']=pd.to_datetime(sales_data['Ship Date'])

#verify the changes
print("\n Data after cleaning: ")
display(sales_data.head())

#Plot Sales trend over time
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(10,6))
sales_data.groupby('Order Date')['Sales'].sum().plot(kind='line', color='purple')
plt.title('Sales trend over time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Scatter plot: Profit vs Discount
plt.figure(figsize=(10,6))
sns.scatterplot(x='Discount', y='Profit', data=sales_data,color='crimson')
plt.title('Profit vs Discount')
plt.xlabel('Profit')
plt.ylabel('Discount')
plt.show()

#Sales distributed by region
plt.figure(figsize=(10,6))
region_sales=sales_data.groupby('Region')['Sales'].sum()
region_sales.plot(kind='bar',color='cyan')
plt.title('Sales by region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()

#Heatmap for correlations
import seaborn as sns
import matplotlib.pyplot as plt
numeric_data = sales_data.select_dtypes(include=['number'])  # Ensuring sales_data contains only numeric values
plt.figure(figsize=(10,6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='PuBuGn')
plt.title('Correlation Matrix')
plt.show()

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

#Select features and target
X=sales_data[['Profit','Discount']]
Y=sales_data['Sales']

#Split the dataset into training and test sets
X_train, X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, random_state=42)

#Train the Linear Regression model
model=LinearRegression()
model.fit(X_train,Y_train)

#Make predictions
Y_pred=model.predict(X_test)

#Evaluate the model
print(f"Mean Squared Error: {mean_squared_error(Y_test,Y_pred):.2f}")
print(f"R.squared Score: {r2_score(Y_test,Y_pred):.2f}")
```
